```python
import pandas as pd
import json
```

```python
df_movies = pd.read_csv('data/movies_dataset.csv')
df_movies
```

```python
df_credits = pd.read_csv('data/credits.csv')
df_credits
```

```python
#Cambiamos el tipo de dato de las ids de los dos archivos
df_movies['id'] = df_movies['id'].astype(str)
df_credits['id'] = df_credits['id'].astype(str)
```

```python
#Unimos los dos datasets con la función merge
data = pd.merge(df_movies, df_credits, on='id')
```

```python
#Borramos índices duplicados
data.drop_duplicates(subset="id", inplace=True)
print(len(data))
```

Creamos una función para desanidar los valores de la columna "belongs_to_collection".

```python
# Definimos una función para desanidar los valores
def desanidar(valor):
    try:
        valor = json.loads(valor.replace("'", '"'))
        return valor
    except:
        return {}
```

```python

# Desanida la columna "belongs_to_collection"
data['belongs_to_collection'] = data['belongs_to_collection'].apply(desanidar)
data['belongs_to_collection']
```

Creamos una función para desanidar los valores de la columna 'production_companies'.

```python
# Define una función para desanidar los valores
def desanidar(valor):
    try:
        valor = json.loads(valor.replace("'", '"'))
        return valor
    except:
        return []

# Desanida la columna "production_companies"
data['production_companies'] = data['production_companies'].apply(desanidar)

# Crea columnas "company_name" e "id_company" y asigna los valores correspondientes
data['company_name'] = data['production_companies'].apply(lambda x: x[0]['name'] if len(x) > 0 else None)
data['id_company'] = data['production_companies'].apply(lambda x: x[0]['id'] if len(x) > 0 else None)

```

```python
data[['production_companies', 'company_name','id_company']]
```

Creamos una función para desanidar los valores de la columna 'production_countries'.

```python
# Define una función para desanidar los valores
def desanidar(valor):
    try:
        valor = json.loads(valor.replace("'", '"'))
        return valor
    except:
        return []

# Desanida la columna "production_countries"
data['production_countries'] = data['production_countries'].apply(desanidar)

# Crea columnas "iso_3166_1" y "country_production" y asigna los valores correspondientes
data['iso_3166_1'] = data['production_countries'].apply(lambda x: x[0]['iso_3166_1'] if isinstance(x, list) and len(x) > 0 else None)
data['country_production'] = data['production_countries'].apply(lambda x: x[0]['name'] if isinstance(x, list) and len(x) > 0 else None)
```

Creamos una función para desanidar los valores de la columna "spoken_languages"

```python
data['spoken_languages'] = data['spoken_languages'].apply(desanidar)

data['iso_spoken'] = data['spoken_languages'].apply(lambda x: [genre['iso_639_1'] for genre in x])
data['language_spoken'] = data['spoken_languages'].apply(lambda x: [genre['name'] for genre in x])
```

```python
data[['iso_spoken', 'language_spoken']]
```

Creamos una función para desanidar los valores de la columna "genres"

```python
data['genres'] = data['genres'].apply(desanidar)

data['id_genre'] = data['genres'].apply(lambda x: [genre['id'] for genre in x])
data['name_genre'] = data['genres'].apply(lambda x: [genre['name'] for genre in x])
```

```python
#Creamos un diccionario

genres_dict = {}
for item in data['genres']:
    for genre in item:
        genres_dict[genre['id']] = genre['name'].lower()
```

```python
genres_dict
```

Creamos una función para desanidar los valores de la columna "cast".

```python
# Desanidar la columna "cast"
def desanidar(valor):
    try:
        valor = eval(valor)
        return valor
    except:
        return []

data['cast'] = data['cast'].apply(desanidar)

# Crear las columnas adicionales
data['id_actor'] = data['cast'].apply(lambda x: [actor['id'] for actor in x])
data['name_actor'] = data['cast'].apply(lambda x: [actor['name'] for actor in x])
```

```python
#Diccionario con la lista de nombres de actores
characters_dict = {}
for item in data['cast']:
    for item in item:
        characters_dict[item['id']] = item['name'].lower()
```

```python
characters_dict
```

Creamos una función para desanidar los valores de la columna "crew".

```python
data["crew"]
```

```python
# Se crea el diccionario de directores
directores_dict = {}
```

```python
# Convertir a lista de diccionarios el data crew
data['crew'] = data['crew'].apply(lambda x: eval(x) if pd.notnull(x) else [])
```

```python
id_directores = []

# Crear las columnas adicionales y desanidar la columna "crew"
for lista in data['crew']:
    directores = []
    for item in lista:
        if item['job'] == 'Director':
            directores_dict[item['id']] = item['name'].strip().lower() # Se agrega el nombre y el id a un diccionario de directores
            directores.append(item['id'])

    # Se añade la lista de directores
    id_directores.append(directores)

data['id_directores'] = id_directores
```

```python
#Los valores nulos de los campos "revenue" y "budget" son rellenados con el número "0"
data['revenue'] = data['revenue'].fillna(0)
data['budget'] = data['budget'].fillna(0)
data['runtime'] = data['runtime'].fillna(0)
```

```python
#Eliminamos los valores nulos del campo "release_date".
data = data.dropna(subset=['release_date'])
```

```python

data.dropna(subset='title', inplace=True)
data['release_date'] = pd.to_datetime(data['release_date'])
data['release_year'] = data['release_date'].dt.year
data['release_month'] = data['release_date'].dt.month
```

```python
dia_español = {
        0: 'lunes',
        1: 'martes',
        2: 'miercoles',
        3: 'jueves',
        4: 'viernes',
        5: 'sabado',
        6: 'domingo'
    }
```

```python
# Se crea columna de día
import calendar

data['release_day'] = data['release_date'].apply(lambda x: calendar.weekday(x.year, x.month, x.day))
data['release_day']
```

```python
def numero_a_dia(num):
    return dia_español[num]
```

```python
# Pasamos el release day a texto por practicidad de la API
data['release_day'] = data['release_day'].apply(numero_a_dia)
data['release_day']
```
